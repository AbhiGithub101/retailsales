```python
import pandas as pd   
import matplotlib.pyplot as plt    
```

```python
df = pd.read_excel('amazon.xlsx')
df
```

```python
ndf = df['Customer Type'].value_counts().reset_index()
ndf
```

```python
plt.pie(ndf['count'],labels=ndf['Customer Type'],autopct='%.1f%%',colors=['green','blue','red'])


plt.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Pie(
    labels=ndf['Customer Type'],
    values=ndf['count'],
    textinfo='percent',
    hoverinfo='label+percent',
    marker=dict(colors=['green', 'blue', 'red'])
)])

fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(showlegend=False)

fig.show()
```

```python
ndf = df.groupby('Product').agg(
    total = ('Quantity','sum')
).reset_index()


l = []

for i in ndf['total']:
    if(i>=500):
        l.append('green')
    else:
        l.append('red')

plt.figure(figsize=(10,5))

plt.bar(ndf['Product'],ndf['total'],color=l)  

plt.xticks(rotation=45)

for m,n in zip(ndf['Product'],ndf['total']):
    plt.text(x=m,y=n,s=n,ha='center',va='bottom')

plt.grid()
plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Year'] = df['Order Date'].dt.year   
df
```

```python
ndf
```

```python
ndf = df.groupby('Year').agg(
    total_sales = ('Sales','sum'),
    total_profit = ('Profit','sum')
).reset_index()


plt.figure(figsize=(12,6))
ndf['Year'] = ndf['Year'].astype(str)
plt.plot(ndf['Year'],ndf['total_sales'],marker='o',label='Sales')
plt.plot(ndf['Year'],ndf['total_profit'],marker='o',label='Profit')

for m,n in zip(ndf['Year'],ndf['total_sales']):
    plt.text(x=m,y=n,s=round(n),ha='center',va='bottom',color='green')

for m,n in zip(ndf['Year'],ndf['total_profit']):
    plt.text(x=m,y=n,s=round(n),ha='center',va='bottom',color='green')

plt.grid()
plt.legend()
plt.show()
```

```python
ndf = df.groupby('Country').agg(
    total_sales =('Profit','sum')
).reset_index()

plt.figure(figsize=(15,5))

ndf = ndf.sort_values(by='total_sales',ascending=False)   
plt.bar(ndf['Country'],ndf['total_sales'])


plt.xticks(rotation=90)
plt.show()
```
